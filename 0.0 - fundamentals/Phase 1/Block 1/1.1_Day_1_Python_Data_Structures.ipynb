## Lernziel
Ich kann `list`, `tuple`, `set`, `dict` erklären und korrekt einsetzen – mit Begründung.

**Merksatz (Intuition):**
- **list** = veränderbare Reihenfolge (Playlist)
- **tuple** = fixe Reihenfolge (Koordinaten, „so ist es“)
- **set** = einzigartige Elemente (Duplikate raus)
- **dict** = Zuordnung (Schlüssel → Wert), wie ein Nachschlagewerk


```python
examples = [
    [1,2,3],            #list
    (48.2082, 16.3738), #tuple
    {1, 2, 2, 3},       #set
    {"C001": 120.0}     #dict
]

for e in examples:
    print(e, "->", type(e))
```

```python
# ZELLE 2 – Liste mit Zahlen erstellen

# 1️⃣ Erstelle eine Liste mit 5 Zahlen deiner Wahl
numbers = [1, 2, 3, 4, 5]  

# 2️⃣ Gib die Liste aus
print("Liste:", numbers)

# 3️⃣ Gib den Typ aus
print("Typ:", type(numbers))

# 4️⃣ Füge eine weitere Zahl hinzu
numbers.append(6) 

print("Nach Hinzufügen:", numbers)


# 5️⃣ Beantworte als Kommentar:
# Warum ist eine list hier sinnvoll?
# Deine Antwort: 
# Eine Liste ist hier sinnvoll, weil sie eine geordnete Sammlung von Elementen ermöglicht, die veränderbar ist.
```

```python
# ZELLE 3 – Tuple mit Koordinaten

# 1️⃣ Erstelle ein Tuple mit zwei Koordinaten (Breite, Länge)
coordinates = (48.2082, 16.3738)

print("Koordinaten:", coordinates)
print("Typ:", type(coordinates))


# 2️⃣ Versuche testweise, den ersten Wert zu ändern
# coordinates[0] = 0  # optional testen


# 3️⃣ Beantworte als Kommentar:
# Warum ist ein tuple sinnvoller als eine list für Koordinaten?
# Deine Antwort:
# Ein Tuple ist sinnvoller als eine Liste für Koordinaten, weil es unveränderlich ist. 
# Koordinaten sollten nicht verändert werden, daher bietet ein Tuple eine bessere Datenstruktur, 
# um die Integrität der Daten zu gewährleisten.
```

```python
# ZELLE 4 – Set mit doppelten Werten

# 1️⃣ Erstelle ein Set mit mehrfach vorkommenden Zahlen
numbers_set = {1, 2, 2, 3, 4, 4, 5}  

print("Set:", numbers_set)
print("Typ:", type(numbers_set))


# 2️⃣ Versuche, auf das erste Element per Index zuzugreifen
#numbers_set[0]   # optional testen


# 3️⃣ Beantworte als Kommentar:
# Wann ist ein set besser als eine list?
# Deine Antwort:
# Ein Set ist besser als eine Liste, wenn du eine Sammlung von einzigartigen Elementen benötigst 
# und die Reihenfolge der Elemente nicht wichtig ist.
```

```python
# ZELLE 5 – Dict mit customer_id und revenue

# 1️⃣ Erstelle ein Dictionary mit mindestens 3 Kunden
customer_revenue = {
    "C001": 120.0,
    "C002": 150.0,
    "C003": 200.0
}

print("Dictionary:", customer_revenue)
print("Typ:", type(customer_revenue))


# 2️⃣ Greife auf den Umsatz eines bestimmten Kunden zu
print(customer_revenue["C001"])  # ggf. anpassen


# 3️⃣ Beantworte als Kommentar:
# Warum ist ein Dict besser als zwei parallele Listen?
# Deine Antwort:
#Ein Dict ist besser als zwei parallele Listen, weil es eine klare Zuordnung zwischen Schlüssel und Wert ermöglicht.
# Es verbessert die Lesbarkeit und Wartbarkeit des Codes, da du direkt auf die Werte über die Schlüssel zugreifen kannst, 
# anstatt die Indizes in parallelen Listen zu verwalten.
```

```python
# ZELLE 6 – Iteration

numbers = [10, 20, 30]

customer_revenue = {
    "C001": 120.0,
    "C002": 150.0,
    "C003": 200.0
}

print("---- Liste iterieren ----")
# 1️⃣ Iteriere über die Liste und gib jedes Element aus
for num in numbers:
    print(num)

print("---- dict.keys() ----")
# 2️⃣ Iteriere über die Keys
for key in customer_revenue.keys():
    print(key)

print("---- dict.values() ----")
# 3️⃣ Iteriere über die Values
for value in customer_revenue.values():
    print(value)

print("---- dict.items() ----")
# 4️⃣ Iteriere über Key und Value gleichzeitig
for key, value in customer_revenue.items():
    print(f"Key: {key}, Value: {value}")
```

# Kontrollfragen – Tag 1

## 1️⃣ Warum ist ein Dict besser als zwei parallele Listen?

Antwort: Ein Dict speichert die Beziehung zwischen Schlüssel und Wert explizit.
Bei parallelen Listen besteht die Gefahr, dass die Index-Synchronisation bricht und falsche Zuordnungen entstehen.


## 2️⃣ Was passiert, wenn ich versuche, ein Set zu indexieren?

Antwort: Ein Set ist keine geordnete Sequenz. Es besitzt keine festen Indexpositionen. Deshalb ist es nicht subscriptable.


## 3️⃣ Wann würdest du absichtlich kein Dict verwenden?

Antwort: Wenn keine Schlüssel-Wert-Beziehung notwendig ist und eine einfache Sequenz ausreicht, z. B. eine geordnete Liste von Messwerten.


# Ein-Satz-Erklärungen

list: Eine geordnete, veränderbare Sequenz von Elementen mit Indexzugrif
  
tuple: Eine geordnete, unveränderliche Sequenz von Elementen.

set: Eine ungeordnete Sammlung eindeutiger Elemente ohne Indexzugriff.

dict: Eine Mapping-Struktur, die eindeutige Schlüssel mit Werten verknüpft und schnellen Zugriff über den Schlüssel ermöglicht.
