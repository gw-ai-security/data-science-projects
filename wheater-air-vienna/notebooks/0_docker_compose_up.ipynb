# Docker‑Stack starten
Wir verwenden eine lokale Docker‑Compose‑Umgebung:

- Prüfung, ob Docker verfügbar ist
- Prüfung, ob docker-compose.yml existiert
- Starten oder Erstellen des gesamten Compose‑Stacks
- Ausgabe hilfreicher Statusmeldungen

Wichtig:
Diese Zelle funktioniert nur, wenn Docker lokal installiert ist und ausgeführt werden darf.

```python
import subprocess
import os

compose_file = "../docker-compose.yml"

def run_command(command):
    """Führt einen Shell-Befehl aus und gibt Erfolg, stdout und stderr zurück."""
    result = subprocess.run(
        command,
        shell=True,
        capture_output=True,
        text=True
    )
    return result.returncode == 0, result.stdout.strip(), result.stderr.strip()

# 1. Prüfen, ob Docker verfügbar ist
ok, stdout, stderr = run_command("docker --version")

if not ok:
    print("❌ Docker ist nicht verfügbar.")
    print("Bitte Docker Desktop installieren und starten.")
else:
    print(f"✅ Docker gefunden: {stdout}")

    # 2. Prüfen, ob docker-compose verfügbar ist
    ok, stdout, stderr = run_command("docker compose version")
    if not ok:
        print("❌ 'docker compose' ist nicht verfügbar.")
        print("Bitte sicherstellen, dass Docker Desktop aktuell ist.")
    else:
        print(f"✅ docker compose gefunden: {stdout}")

        # 3. Prüfen, ob docker-compose.yml existiert
        if not os.path.exists(compose_file):
            print(f"❌ '{compose_file}' wurde nicht gefunden.")
            print("Bitte sicherstellen, dass die Datei im Projektordner liegt.")
        else:
            print(f"✅ '{compose_file}' gefunden.")

            # 4. Prüfen, ob der Stack bereits läuft
            ok, stdout, stderr = run_command("docker compose ps --format json")

            stack_running = ok and stdout.strip() != ""

            if stack_running:
                print("docker-compose Stack existiert bereits → wird gestartet …")
                ok, stdout, stderr = run_command("docker compose up -d")
                if ok:
                    print("✅ Stack wurde erfolgreich gestartet.")
                else:
                    print("❌ Stack konnte nicht gestartet werden.")
                    print(stderr)
            else:
                print("docker-compose Stack existiert noch nicht → wird erstellt …")
                ok, stdout, stderr = run_command("docker compose up -d")
                if ok:
                    print("✅ Neuer docker-compose Stack wurde erstellt und gestartet.")
                else:
                    print("❌ Stack konnte nicht erstellt werden.")
                    print(stderr)
```
